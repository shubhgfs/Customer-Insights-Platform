```python
model_name = os.getenv("AZURE_OPENAI_DEPLOYMENT_ID_AQMAGENTICOS")
azure_openai_api_key = os.environ["AZURE_OPENAI_API_KEY_AQMAGENTICOS"]
azure_openai_endpoint = os.environ["AZURE_OPENAI_ENDPOINT_AQMAGENTICOS"]
llm = AzureChatOpenAI(
    openai_api_version=os.getenv("AZURE_OPENAI_API_VERSION_AQMAGENTICOS"),
    azure_deployment=model_name,
    model_name=model_name,
    temperature=0.0)
httpx_client = httpx.Client(verify=False)
```

```python
from langchain.chains import create_sql_query_chain
chain = create_sql_query_chain(
    llm=llm,
    db=db
)
# Make sure to use a client with SSL verification disabled for the chain
response = chain.invoke(
	{"question": "What is the conversion rate for Jan month ?"},
	config={"callbacks": None},
	httpx_client=httpx_client
)
print(response)
```
